### 1. Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page

```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
```

```python
# 2. url: we start with the 'second' page. Show that you can start whenever you want
url ='https://en.wikipedia.org/wiki/Python'
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
soup
```

```python
soup.select("#bodyContent")
```

```python
linklist =[]
num_iter = len(soup.select("#mw-content-text > div.mw-parser-output")[0].find_all('a',href=True))
for i in range(num_iter):
    link = soup.select("#mw-content-text > div.mw-parser-output")[0].find_all('a',href=True)[i]
    linklist.append(link['href'])
```

```python
len(soup.select("#mw-content-text > div.mw-parser-output")[0].find_all('a',href=True))
```

```python
soup.select("#mw-content-text > div.mw-parser-output")[0].find_all('a',href=True)[3]["href"]
```

```python
linklist
```

### 2. Find the number of titles that have changed in the United States Code since its last release point

```python
# 2. url: we start with the 'second' page. Show that you can start whenever you want
url ='http://uscode.house.gov/download/download.shtml#fn'
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# 4.2. check that the html code looks like it should
soup
```

```python
soup.select(" div > div.uscitemlist")[0].get_text()
```

```python
title =[]
num_iter = len(soup.select(" div > div.uscitemlist"))
for i in range(num_iter):
    link = soup.select(" div > div.uscitemlist")[0].find_all("div",{"class": "usctitlechanged"})[i].get_text()
    title.append(link)
```

```python
print (len(title))
```
